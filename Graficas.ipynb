```julia
R1=Interval(900,1100)
R2=Interval(900,1100)
R3=Interval(900,1100)

V1=Interval(10)
V2=Interval(5)
A=[R1+R2 -R2;-R2 R2+R3]
b=[V1;-V2];

x = [1;2;3;4;5;6;7;8]
y = zeros(length(x))
z = zeros(length(x))
v = zeros(length(x))
w = zeros(length(x))
for i in x
    y[i] = KrawczykMethod(A,b,i)[1].right
    z[i] = KrawczykMethod(A,b,i)[1].left
    v[i] = KrawczykMethod(A,b,i)[2].right
    w[i] = KrawczykMethod(A,b,i)[2].left
end
p_x=plot(x, y, x, z)
p_y=plot(x, w, x, v)
```
